```python
#라이브러리 임포트는 항상 처음부터 해주세용~
import pandas as pd                     #데이터 컨트롤
import numpy as np                      #더미 데이터 생성
import matplotlib.pyplot as plt         #시각화. pyplot: 파이썬으로 플롯을 그리겠다. 맷플롯립의 여러 버전 중 하나.
```

#1. 선형 그래프
- 일차 방정식

```python
#수동 더미 데이터 입력
#주의: 각 변수간 리스트 내 원소 개수(=길이) 동일하게
x=[1,2,3,4,5]
y=[2,6,7,8,10]

#선형 그래프 생성

plt.plot(x,y)

#플롯과 쇼 사이에 그래프 보조지표 등 꾸미기 설정 위치.
#그래프 제목 설정
plt.title('My First Plot')

#x,y축 이름 설정
plt.xlabel('month')
plt.ylabel('score')

plt.show()          #show() = print() 같은 역할.

#plt.savefig('/content/figure.png')

#./~와 /~ 의 차이
#/ ('~ 드라이브부터 시작하자'라는 뜻) 절대 경로: 지정된 직접적인 위치, C드라이브 먼저 처음부터 모든 위치를 설명해야함.
#./ ('지금 여기 밑에' 라는 뜻) 상대 경로: 간접 위치. 지금 이 파일을 기준으로~ 시작함.
```

```python
#맷플롯립 도화지 설명 (그리는 방법, 기본적인 원리)

#도화지 정의, 접어서 어떤 방을 어떻게 쓸건지~
#fig = 도화지 그 자체   -> 항상 1개                     도화지 전체 설정은 fig.
#ax = 그림              -> 선택에 따라 여러 개          그림 개별 설정은 ax.
                    # 섭플롯(figsize(가로,세로)) = 도화지 크기
fig, ax = plt.subplots(figsize=(8,6))                   #도화지 설정( 섭플롯이 여러 그래프를 연결하는 경우 편함 )

#그림 그리는데, 선형 그래프고, (x.y) 데이터 넣어 그래프를 그림.
ax.plot(x,y)

plt.show()
```

```python
#더미 데이터 생성. numpy
#(시작, 끝, 몇 개의 데이터): 시작부터 끝까지 몇개의 데이터로 나누거라
x = np.linspace(0,10,100)
```

```python
#plt.subplots(행,열, figsize=(가로,세로))
fig, ax = plt.subplots(2,1,figsize=(10,15))

#그래프가 2개 이상이면 ax는 리스트가 됨.

ax[0].plot(x,x**2)  #제곱

ax[1].plot(x, x+1)


plt.show()

```

```python
#그림의 특정 요소 바꾸기

x= np.linspace(0,30,10)
y= x*2


#1행 1열 (figsize= 만 필요할 듯)
fig, ax = plt.subplots(figsize= (4,4))

#선형 그래프 옵션 설정
ax.plot(x, y,                           #구글링해서 옵션 찾으시오
        linestyle = '-.',           #'--'와 같이 직관적인 스타일로 가능
        marker = "s",
        color = 'navy', )           #헥사코드사용 가능


plt.show()
```

```python
fig, ax = plt.subplots()                                #도화지 설정
                                                                                                                            #그림도 한 장

x = np.arange(10)                                      #x는 더미데이터, 범위? 수열? 뭐더라 10 설정.                        #range(10) 과 동일함
        #(x축 데이터, y축 데이터, 기타 옵션~)
ax.plot(x, x, color='red', label='1*ratio')             #x=x그래프, 선 색은 빨간색, 제목은 '1배율'                              #label: 그림에 설명을 붙임(범례). .legend와 연결됨(범례가 있어야 legend 생성).
ax.plot(x, x**2, color='green', label = 'squared')      #x=x제곱 그래프, 선 색은 초록색, 제목은 '스퀘어드 =??'              #한 그래프에 여러 선 설정한 것
ax.plot(x, x*1.5, color='blue', label = '1.5*ratio')    #x=1.5배율 그래프, 선 색은 파란색, 제목은 1.5배율

ax.legend(loc='upper right')                            #레전드 뭐임.ㄹㅈㄷ 해석 불가 ㄹㅈㄷ 로케이션 = '대문자 롸잇 얜 또 몰가     #범례 위치 설정. 우측 상단에 설정한다는 뜻


plt.show()
```

## -2.  산점도 (scatter)


```python
fig, ax = plt.subplots()

x = np.arange(10)
                        # 점 간 간격 줄이기는 어렵당 그림 크기를 줄이면 가능
ax.scatter(x, x*2, marker = '*', color = 'blue')

plt.show()
```

```python
fig, ax = plt.subplots()

x = np.arange(10)
                        # 점 간 간격 줄이기는 어렵당 그림 크기를 줄이면 가능
ax.bar(x, x*2, color = 'blue')

plt.show()
```

```python
# 값 누적된 막대 그래프

#1. 가상 데이터 생성
                            # 정규 분포표 내 가상 숫자 사용
x = np.random.rand(3)       #괄호 내 숫자 기입하면 자동 랜덤 생성
y = np.random.rand(3)
z = np.random.rand(3)


data = [x, y, z]            #여기까지 y 데이터
x_ax = np.arange(3)         #세 개의 값을 임의로 함.x 데이터는 0,1,2 고정임, (arange니까)


#2. 그래프 생성

fig, ax = plt.subplots()

for i in x_ax:                            # 0,1,2가 돌아가며
                                          # 막대 그래프 만들거야(x 데이터(x_ax), y 데이터(data[i])=x,y,z 돌아가며 들어감)
    ax.bar(x_ax, data[i],                                     #axis = 0이면 행 방향, 1이면 열 방향
        bottom = np.sum(data[:i], axis=0) )                   #bottom = 막대 누적 방향. data [:i]: 데이터를 i까지 쌓아라

#x축 이름 변경 (간격?)

# ax.set_xticks([0,1,2])        #fig.xticks(~) 하면 도화지엔 그대로 들어가는데, 그림에는 추가로 설정이 필요. ax.set_xticks()
    #결과 0,1,2

ax.set_xticklabels(['A', 'B', 'C'])
    #결과 A, B, C

plt.show()
```
