```python
data = pd.DataFrame([[1, 2, 3],
                     [2.3, np.nan, 4],
                     [np.nan, np.nan, np.nan],
                     [np.nan, 44, np.nan]])


print(data)
```

```python
# 모수적 방법

#1. 평균값 대체

x = data[0].mean()
data.fillna(value=x)     #특정 열의 빈 곳을 '평균'으로 대체

#2. 중앙값 대체

```

```python
#fillna 는 들어갈 값이 필수이기에 value 는 꼭 넣을 것
data.fillna(value = 0,axis=0)
```

```python
#method 옵션: '어떻게 채울 것인가' 에 대한 것. 즉 value가 없어도 가능.
data.fillna(method='ffill')             #foward fill(선행값으로 채우기), back fill(후행 값으로 채우기)
                                #이건 그럼 axis 적용되나?

```

```python
data.fillna(method='ffill', axis=1)                     #이 옵션에서 행과 열 방향 설정해도 적용이 되는것이냐
```

```python
#IQR 방식 진행

#1. q1~q3 구해야함 (IQR = Q3 - Q1 구하기)
#2. q1 - 1.5IQR 을 낮은 기준으로
#3. Q3 + 1.5IQR 을 높은 기준으로
#4.모든 데이터를 돌며 3보다 작거나 4보다 큰지 알아보기
#5. 5의 조건에 해당하면 이상치(수염범위)

def iqr(col) :             #왜 함수로 만들게요?: 데이터프레임의 특성이라,  열별(e.g. 점수)로 이상치 탐지해야함,
    #1. q1~q3 구하기
    Q1 = col.quantile(0.25)
    Q2 = col.quantile(0.5)          #col.median()
    Q3 = col.quantile(0.75)
    #IQR 구하기
    IQR = Q3 - Q1

    #2, 3
    lower_band = Q1 - 1.5*IQR
    upper_band = Q3 + 1.5*IQR

    outliers = []
    #4.
    for i in col:                   #열 안에 있는 각 행이 i가 됩니다
        if i < lower_band or i > upper_band:
            outliers.append(i)
    return outliers
```
