```python
## replace: 특정 데이터를 다른 값으로 대체함

data = {"Peter" : "Man",
        "Sally" : "Woman",
        "Chris" : "Woman",
        "Jack" : "Man"  }

human = pd.DataFrame(data, index =[0])      #index =[0]: 인덱스를 0번으로 넣은 것 이름을 보존하려 더미 데이터 삽입. 결과 보면 알거임.
```

```python
human = human.T
```

```python
human.columns=['Gender']
```

```python
human
```

```python
#M,W로 바꾸기
#human['Gender'].replace() : 원래하던 방식.
#'Man' -> 'M'
#'Woman' -> 'W'
# 바꿀 내용이 많으면 딕셔너리 활용
human.Gender.replace({"Man" : "M", "Woman" : "W"}, inplace = True)

#inplace를 써야 human 불러야함
human
```
