```python
#1. 정답단어를 고른다.
#초기화
result = 'aggregate'        #정답단어
count = 5                   #생명개수
memory =[]                  #사용자가 말한 알파벳을 저장하는 리스트
guess = ['_' for x in range(len(result))]                    #사용자에게 보여줄 대답 내용. 얘는 가로로 쭉 나옴



#반복 조건:
#1. 생명이 남아있는가?
#2. 맞출 것이 남아 있는가?

while count > 0 and '_' in guess:                    #guess isalpha ? 써도 되나용

    letter = input('알파벳 하나를 입력하시오.\n')
    memory.extend(letter)               #extend: 추가. append와 차이점은
#예외처리
    #1. 두 글자 이상 쓴 경우
    if len(letter) >2:
        print ('한 글자만 입력하시오.')


    #2. 공백 입력한 경우
    #'a'isalpha(): 'a'가 알파벳인지 판별하고 T/F 도출
    if letter.isalpha()==False:
        print('알파벳이 아닙니다.')

        # if letter == ' ':
        #     print()? 도 되려나?

    #3. 같은 글자를 또 쓴 경우
    #1) 이전 입력값을 저장해야함 = 복수 자료형을 이용. 리스트가 간편하니 리스트로.
        #여기에 memory=[] 하면 얘 건망증 온다
    #2) 사용자가 말한 글자가 저장된 메모리 안에 있는지 확인.
    if letter in memory:
        print('이미 대답한 글자입니다.')


#3. 2에서 사용자가 고른 알파벳이 정답 단어 안에 있는지 확인한다.
    if letter in result:
        print('맞췄습니다!')
            for i,v in enumerate(result):   #if 있다면 -> 어느 위치에 알파벳이 있는지 적어서 보여준다. #필요한 것: 어느 위치에 어떤 알파벳. = 알파벳의 위치 구해야해요
                if letter == v:
                    print(v)
                    guess[i] = letter       #guess의 i번째 위치를 letter로 지정.
                else:
                    print('_')
    else:                                       #if 없다면 -> 생명이 -1
        count -= 1                          #얘는 for문 밖으로 보내야함. 왜? 동작하는 만큼 (단어 1회독)생명이 깎이는 꼴.

    print(guess)                            #사용자의 대답 기록을 표현함.

    if '_' not in guess:
        print(f'성공했습니다. 정답은 {result}입니다.')
        break
```

기능 요구사항 명세서를 확실히 받아야함.
기능. 기능 요구사항. 예외사항 처리 등 작업 전체 플로우를 이해할 것.

```python
#행맨 쌤 코드

#초기화
result = 'aggregate' #정답단어
count = 10           #생명개수
memory = []          #사용자가 말한 알파벳을 저장하는 리스트
guess = ['_' for x in range(len(result))] #사용자에게 보여줄 대답 내용


#반복 조건:
#1. 생명이 남아있는가?
#2. 맞출것이 남아있는가?
while count > 0 and '_' in guess:
    letter = input("알파벳 하나를 입력하시오.\n")
    memory.extend(letter)

    #1.두글자 이상 쓴 경우
    if len(letter) >= 2:
        print('한글자만 입력하시오.')


    #2.알파벳이 아닌것을 입력한 경우
    if letter.isalpha() == False :
        print('알파벳이 아닙니다.')


    #3.같은 글자를 또 쓴 경우 ->
    #3-1. 사용자가 말한 글자를 저장해야 함. (위에 memory로 만듦)
    #3-2. 사용자가 말한 글자가 저장된 memory 안에 있는지 확인해야 함
    if letter in memory:
        print('이미 대답한 글자입니다.')


    #3.2에서 사용자가 부른 알파벳이
    #정답 단어 안에 있는지 확인한다.
    if letter in result:
        print('맞췄습니다!')
        for i, v in enumerate(result):
            if letter == v:
                guess[i] = letter


    else:
        count -= 1
    print(guess) #사용자의 대답 기록을 표현


    if '_' not in guess:
        print(f'성공했습니다! 정답은 {result}입니다.')
        break
```
