```python
dictA = {'3':4, '5':6}
listA = [1,2,3,4]


#enumerate(): 함수. 데이터 순서와 리스트 내의 값을 동시에 출력함. 순서와 리스트 값이 필요할 때.

for i in enumerate(listA):      #순서와 리스트 내의 값을 (순서,리스트 값)으로 출력.
  print(i)

for i,v in enumerate(listA):      #순서 --> 리스트내 값
  print(i, '--->', v)

#위의 예제는 리스트와 함께 쓴 이너멀 함수형
```

```python
#이너멀: 키와 값을 쌍으로 인쇄
for i,v in enumerate(dictA.items()):      #순서 --> 리스트내 값
  print(i, '--->', v)

```

```python
#결과 쪼개기: zip(지퍼 형태)
#[1,2,3,4], [5,6,7,8] 같은 위치의 원소들을 묶습니다
for i,v in zip(dictA.keys(), dictA.values()):
  print(i, '--->', v)


  # [3, 5]  [4, 6], 두 리스트 중 첫번째 자리인 3과 4, 둘째 자리인 5,6을 묶어서 도출.
```

```python
# while문 하나 틀리면 아예 껏키다 조심해라
# while (조건식):
count = 0
while (count<10):
  print(count)
  count += 1        #count = count+1 할당연산자. (등호 앞 다양한 연산자 삽입 가능)
```

```python
# #과제1
# #1부터 30까지의 숫자 반복. 다음 규칙 따라 숫자 출력
# 3의 배수: "Fizz"
# 5의 배수: "Buzz"
# 3과 5의 공배수: "FizzBuzz"
# 그 외: 숫자 그대로
# 반복, 조건문 사용


for i in range (1,31):
    if i %3==0 and i%5==0:
        print("Fizz")
    elif i %5==0:
        print("Buzz")
    elif i %3==0:
        print("FizzBuzz")
    else:
        print(i)
```

내가 놓친 점 찾으시오
- 조건 중에 덩어리가 크고 정보가 큰 조건을 우선으로 작성해줘야함 (fizzbuzz 조건문)
- 또한 for while의 구분이 흐릿함 for문이 적합한 문제였음(길이가 정해짐)

```python
#과2
# number1과 number2를 받고, 넘1부터 넘2까지의 합을 구해 출력.
# 단, 넘1 < 넘2
# e.g. number1 = -3, number2 = 5, 결과물은 -3에서 5까지의 합을 구해라
#사용 문법: input, for, 할당연산자(+=꼴)


num1 = input('첫번째 숫자를 입력하세요\n')
num2 = input('두번째 숫자를 입력하세요\n')

num1 = int(num1)
num2 = int(num2)


add = 0                     #누적되는 값 변수 = 0, 먼저 초기화 작업임
for i in range(num1,num2+1):
    add += i                #add = add + i

print('결과는', add, '입니다.')
#F포매팅
print(f'결과는 {add} 입니다.')
```

*조건살리기

```python
num1 = input('첫번째 숫자를 입력하세요\n')
num2 = input('두번째 숫자를 입력하세요\n')

num1 = int(num1)
num2 = int(num2)

if num1 < num2:

    add = 0                     #누적되는 값 변수 = 0, 먼저 초기화 작업임
    for i in range(num1,num2+1):
        add += i                #add = add + i

    print('결과는', add, '입니다.')
    #F포매팅
    print(f'결과는 {add} 입니다.')
else:
    print('조건 맞지 않아 수행 불가입니다.')
```

```python
#과3
#while문 사용해 10부터 카운트 출력, 0초가 되면 '발사!'라고 출력.
#while 조건이 붙는 반복

count = 10
while count >= 0:

    if count == 0:
        print('발사!')
    else:
    #카운트 인쇄 역할
        print(count)

    #카운트 감소 역할
    count -= 1

```
