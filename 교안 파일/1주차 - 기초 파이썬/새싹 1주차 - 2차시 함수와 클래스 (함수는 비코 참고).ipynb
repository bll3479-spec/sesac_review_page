```python
#4. 다음 성적 리스트.
scores = [10, 20, 20, 15, 22, 18, 7, 9]
#10점 미만의 성적을 '재수강으로 분류하는 코드 만들기.
#이 때, 재수강으로 분류된 점수는 list에 추가되어야 함

# for문 활용(while 활용도 가능)

fail=[]

for i in scores:
    if i < 10:
        fail.append(i)
print(fail)


```

```python
#응용
fail = [x for x in scores if x < 10]
print(fail)
```

```python
#class

class bread():
    def __init__(self, order, source):
        self.apperance = order
        self.jam = source
    def bake(self):
        print(f'{self.jam} 맛 {self.apperance}을 구웠습니다.')


b1 = bread('국화빵', '커스타드')        #인자 받기(b1은 클래스 bread이고요, 안에 넣을 인자는 이렇습니다)

b1.bake()                               #bake라고 클래스 bread 내 지정해둔 함수를 실행하세요
```

```python
#클래스 상속

class peanut_bread(bread):                     #땅콩빵이라는 클래스는 bread라는 클래스를 상속.
    def __init__(self):                         #
        super().__init__('땅콩빵', '땅콩잼')

b2 = peanut_bread()
b2.bake()
```
