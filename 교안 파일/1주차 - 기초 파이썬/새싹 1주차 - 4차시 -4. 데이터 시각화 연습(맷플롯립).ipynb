```python
# 날짜와 시간을 분리. 분리한 결과에서 날짜만 사용.
# 날짜에서도 연,월,일로 나누기.
# 연,월,일로 나누는 반복문 만들기.
#     반복문에서 나오는 결과-> 새로운 리스트에 결과값 모두 받아주기
# 이제 다시 한 데이터 프레임으로 만들기
#     그 전에 체크해야할 건 카피본과 분리해낸 연,월,일 리스트가 서로 길이가 같은지 확인
# 체크 결과 이상 없다면 연,월,일 리스트를 데이터 프레임 형식으로 바꿔주기.
# 같은 데이터 프레임 형식이 되었으니 합치기
#     합칠 때 카피본과 연월일 데이터 프레임이 갖고 있는 데이터가 다르니 행과 열 기준을 잘 설정해서 합치기
```

```python
#함수 묶기 (지금은 datesold의 0번째 행이니까 전반적인 저장을 위해)
#def  ():
years = []                                                      #for문 밖에서 리스트 정의해야함 왜? for문은 돌아가며 새로운 리스트를 계속 생성함. 1리스트 1 원소 꼴
months = []
days = []


for i in range(len(sales)):
    x = sales['datesold'][i].split(' ')[0]                      #결과:'2007-02-07'

    x.split('-')                                                #결과: ['2007', '02', '07']

    #x의 0번째~분리
    x = x.split('-')
    year = x[0]
    month = x[1]
    day = x[2]                                                # 다음에 뭐가 올까: 지금 이 상태는 처음부터 끝까지 분리해줌. 연,월,일 일자를 각각 리스트로 받아주는 함수가 필요해~ (왜냐면 난 sales에 연,월,일을 세 열로 만들고 싶거든)

    #리스트.추가(for문 결과값)                                      # 위치 진짜 못 보겟네 아오
    years.append(year)
    months.append(month)
    days.append(day)                                                #이제 sales에 넣어야해 가장 중요한건 열 길이가 똑같아야함!!

    #print(year, month, day)                                    # 중요: 프린트를 for문 내에 넣어야 전체로 프린트 해줌. for문 밖에 있으면 덮어쓰기를 반복하는 꼴이다!!
```

```python
print(years)

len(sales) == len(days)       #이제 sales에 넣어야해 가장 중요한건 열 길이가 똑같은지 확인하는 절차
```

```python
#이제 열을 붙여주자

#리스트 -> 데이터 프레임화
#이중리스트 활용

seperated = pd.DataFrame([years, months, days])
seperated = seperated.T


seperated.columns = ['year', 'month', 'day']
seperated
```

```python
#이제 sales에 seperated 합치기

#pd.concat(): 여러 데이터 프레임을 합칠 수 있음

    #문제 발생: sales 와 sperated는 갖고 있는 데이터가 다르니까 a, n \n n, b 꼴임. (디폴트 axis는 0(행)기준)
    #기준이 필요하다: axis = 0,1

result = pd.concat([sales, seperated], axis = 1)

result
```
