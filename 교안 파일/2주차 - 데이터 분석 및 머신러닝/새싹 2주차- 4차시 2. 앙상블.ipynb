# 앙상블

 - 여러 개의 모델을 같이 쓰는 기법
 최종 대답
 - 배깅: 한 번에 거수. 모델 최종 예측할 때 다수결 결정
 - 부스팅: 1:1 인터뷰. 모델 최종 예측할 때 순서대로. 오류를 발생시킨 데이터에 가중치 부여. (오답이면 가중치 부여해서 다음 질문 모델에 적용)

```python
#AdaBoost: 부스팅 알고리즘
from sklearn.ensemble import AdaBoostClassifier #부스팅

#이하는 다른 모델들을 배깅(), 보팅()으로 감싸서 활용함. 알고리즘이 아니라 추가가 필요. 리스트로 묶어서 넣을 것임
from sklearn.ensemble import BaggingClassifier  #배깅
from sklearn.ensemble import VotingClassifier   #모델들이 다수결

#개별 분류기(약한 분류기)
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
#여러개의 모델을 한꺼번에 돌려서 확인
classifier = [DecisionTreeClassifier(), GaussianNB(), KNeighborsClassifier(), SVC(kernel = 'linear'), SVC(kernel = 'rbf')]      #왜 svc는 두개? 옵션 차이는 아는데 이렇게 써야하는 이유?

#보팅 통해 여러개의 모델 사용 -> 다른 모델을 여러개. 공통의 훈련 자료.
#vote = VotingClassifier(estimators = [ ('이름', 모델 객체()),('이름', 모델 객체())])
vote = VotingClassifier(estimators = [
                            ('svc', SVC()),
                            ('nb', GaussianNB()),
                            ('knn', KNeighborsClassifier())])


#배깅 통해 여러개의 모델 사용 -> 같은 모델을 여러개. 각 모델들이 각자의 훈련자료를 봄.
bag = BaggingClassifier(estimator = SVC(), n_estimators = 10)
```

```python
from sklearn import datasets
data = datasets.load_breast_cancer()
X,Y = data.data, data.target
print(X[0], Y[0])                   #데이터셋에서 첫번째 데이터 - 첫번째 정답 인쇄
```

```python
#voting 훈련

vote.fit(X,Y)

#voting 예측
v_pred = vote.predict(X)

#정확도 평가
from sklearn.metrics import accuracy_score
print(accuracy_score(Y, v_pred))
```

```python
#bagging 훈련

bag.fit(X,Y)

#bagging 예측
b_pred = bag.predict(X)

#정확도 평가

print(accuracy_score (Y, b_pred))
```
