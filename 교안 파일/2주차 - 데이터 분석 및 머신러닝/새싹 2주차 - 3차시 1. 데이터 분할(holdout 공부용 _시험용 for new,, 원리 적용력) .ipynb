# 1. 데이터셋 로드

```python
data = load_wine()

#data 데프로 변환

X = pd.DataFrame(data.data, columns = data.feature_names)

Y = pd.DataFrame(data.target, columns = ['target'])
```

```python
df = pd.concat([X,Y], axis = 1)
df.head()
```

```python
#계층적 K-Fold: K개로 X 데이터를 나누되, 폴드 안에 들어가는 0,1,2 클래스의 비율을 반영해(원 데이터셋의 비율 유지) 나눠줌
```

####1. 홀드 아웃 (train-test 쪼개기)

```python
#train_test_split(X특징, Y정답, test-size=비율, random state = 랜덤 숫자 )

#test-size=비율은 데이터 양에 따라 충분하면 7:3, 적으면 8:2~9:1, 95:5도 가능

#random state: 랜덤으로 데이터셋을 train-test로 분리
#x,y 그리고 train-test 이 받는 순서는 늘 고정 시키시오(변수명은 상관 없음)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2 , random_state = 42)
```

```python
print(f'train 개수 {len(x_train)}, test 개수 {len(x_test)}')
```

```python
x_train         #확인: 왼쪽 인덱스 숫자를 보면 랜덤. 즉, 랜덤으로 뽑힌 train
```

```python
#가우시안 나이브베이즈 적용

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()

nb.fit(x_train, y_train)            #특징 연습용, 정답 연습용으로 훈련

nb_predict = nb.predict(x_test)     #문제지만 제공하는 꼴 = 시험. 문제로 낸 x 테스트를 준 것.
accuracy_score(y_test, nb_predict)  #y_test는 x_test의 답
```

```python
#선호 분류기로 홀드아웃 적용한 생성-훈련-예측-성능평가 써보기

#홀드아웃 적용
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)

#분류기: 의사결정나무 적용
from sklearn.tree import DecisionTreeClassifier

#정의
dt = DecisionTreeClassifier()

#훈련
dt.fit(x_train, y_train)

#예측
dt_predict = dt.predict(x_test)

#평가
accuracy_score(y_test, dt_predict)
```

```python
#홀드아웃 적용
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)

#분류기: 의사결정나무 적용
from sklearn.ensemble import RandomForestClassifier

#정의
rf = RandomForestClassifier()

#훈련
rf.fit(x_train, y_train)

#예측
rf_predict = rf.predict(x_test)

#평가
accuracy_score(y_test, rf_predict)
```

##### KFold 교차검증

- stratifiedKFold
- crossval -> Kfold해서 훈련시킨 모델의 검증(평균으로)

```python
X
```

```python
#선 폴딩 셋팅
#계층별 K폴드 교차 검증
#n_splits= K
#shuffle= 폴드 나눌때 데이터 섞어서 나눌까예?
skf = StratifiedKFold(n_splits=4, shuffle=True)

from sklearn.svm import SVC
svc = SVC()


#cross_val_score(): K(여기선 4) 폴드로 나뉜 데이터를 훈련 시킨 뒤, 평균 성능을 내줘.
svc_result = cross_val_score(estimator = svc,       #모델 알고리즘 넣는 곳 (이 방법으로~)
                X = X,                              #옵션 제목 중 X만 대문자. y는 소문자
                y = Y,                              #애초에 KFold 자체가 데이터를 나누기에 미리 쪼개지 않은 원본, X,Y를 줘도 알아서 해줌
                cv = skf)                           #cv에는 (계층적) k폴드 객체 넣음. 앞에서 말한 skf. Kfold 방식으로 알아서 쪼개시요
```

```python
print(f'정확도: {svc_result.mean():.2f}')       #:.2f => ~의 결과의 소수 둘째자리까지 소수형으로 표현해주삼
```

```python
#1. kfold 교차검증 코드 이용해서 10 폴드로 나눈 훈련 수행하기
#알고리즘은 로지스틱 회귀 사용
#성능 평균까지 도출하시예

# 만약 완전 처음이면(홀드아웃 적용 안된 상태) 홀드아웃부터 적용하는건가?: 맞긴함 skf를 안 할거면 홀드아웃을 적용해주고 해당하는 값을 넣어주면 됨

#알고리즘 적용
from sklearn.linear_model import LogisticRegression
#객체 정의
lr = LogisticRegression()

skf = StratifiedKFold (n_splits = 10, shuffle = True)

lr_result = cross_val_score(estimator = lr, X=X, y=Y, cv = skf)
```

```python
print(f'정확도: {lr_result.mean():.2f}')
print(f'정확도: {lr_result.min():.2f}')
print(f'정확도: {lr_result.max():.2f}')
```
