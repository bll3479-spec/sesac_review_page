```python
import pandas as pd             #행렬데이터를 보고 판단하니꽈

#머신러닝에 대한 알고리즘을 갖고 있는 사이킷런
#라이브러리 중 규모가 큰 것들은 이렇게 아래처럼 일부만 가져올 수 있음
from sklearn.datasets import load_iris  #(load_iris: 이거 함수임)
```

```python
#load_iris(): iris 데이터를 다운로드 받으시오
data = load_iris()


#머신러닝에서 내가 알고 있는 수집 가능한 실제 데이터 값 -> data의 data (data.data)  => x
#머신러닝에서 나의 모델(머신러닝)이 예측하고자 하는 값 -> data.target  => y
print(data.data, '\n')      #꽃잎의 길이, 꽃잎의 넓이,,~ 4가지 항목 (dimensionality = 4)
print(data.target)          #품종
```

```python
data.feature_names
```

```python
#데이터프레임 변환
X = pd.DataFrame(data.data,columns = data.feature_names)
Y = pd.DataFrame(data.target, columns = ['species'])
```

```python
X.head()
```

```python
#concat([여러 개의 데이터 프레임], axis= 축 옵션)
df = pd.concat([X,Y], axis = 1)
df.head()
```

#KNN 알고리즘

 -K Nearest Neighbor: 유유상종

```python
#사이킷런의 knn 알고리즘(분류)을 import 코드
from sklearn.neighbors import KNeighborsClassifier          #사이킷런의 이웃들에서 KN로된 분류기를 가져오시오

knn = KNeighborsClassifier(n_neighbors = 3)                 #클래스.객체 정의. =  KNN 알고리즘 수행할 것을 생성. (이웃의 수는 3명으로 옵션 지정) 얘는 knn이라 정의
#fit은 knn 모델 안에 학습한 흔적이 남도록 설정되어 있음. 별도로 변수 지정 knn=knn.fit(~)하지 않아도 됨
knn.fit(X,Y)                                                #knn에게 X,Y를 인풋, 이를 기반으로 훈련 지시.(모델.fit(문제,정답지)형태)

#정답 안 보고 학습 내용 토대로 '예측' predict
#knn_predict는 예측 결과. 이 결과가 좋으면 시장에서 배포 가능, 결과 안 좋으면 재훈련
knn_predict = knn.predict(X)                                #knn에 X값을 줄테니 예측치를 도출하시오.  #분석가 입장에선 예측치가 얼마나 잘 맞는지만 판단하면 됨.
```

```python
print(knn_predict)
```

```python
#얼마나 잘 맞는지? 판단하기

# 정확도 확인 함수
            #.metrics : 평가기준 폴더~
from sklearn.metrics import accuracy_score
accuracy_score(Y,knn_predict)       #(정답,예측)
```

## 2. 의사결정 나무,  랜덤 포레스트 알고리즘


- 성능이 좋음 쥔짜 좋음


-1. 의사결정 나무
- easy to understand
- 1 알고리즘

-2. 랜덤 포레스트 = 앙상블
- 다수의 의사결정 나무 조합

```python
#의결나무

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X,Y)

dt_predict = dt.predict(X)
accuracy_score(Y, dt_predict)

#단점: 보통 1점이 나온다..
#즉, 과적합 문제 or 과소적합. (문제가 너무 많아!) 보통 과적합이 더 나옴
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score      #성능 평가 안된다? 요놈 추가하삼

#객체 정의                                      #기준
rf = RandomForestClassifier(n_estimators = 50, criterion = 'entropy')           #괄호 내 옵션 삽입 가능, 옵션 검색은 sklearn.ensemble 검색하쇼. n_estimators :나무 몇 그루?(기본 100), max_depth: 나무 가지의 깊이

#훈련
rf.fit(X,Y)

#예측
rf_predict = rf.predict(X)

#성능 평가
accuracy_score(Y,rf_predict)
```
