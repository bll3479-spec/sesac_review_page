```python
import seaborn as sns

tips = sns.load_dataset('tips')
len(tips)
#tips.head()
```

```python
#파생변수 만들기
# 1. 팁/전체 계산 금액 으로 전체 금액 대비 팁의 비율 계산해 tips['ratio']열 생성

#나: 복잡하게 돌아가는듯 일단 자료를 파악하삼, 문제에서 요구하는 것은 수치/수치라 간단히 작성 가능
tips['ratio'] = tips['tip']/tips['total_bill']
tips.head()

#2. '성별'이 'Male'인 경우, '남, 'Female'인 경우 '여'로 적용.

#replace(조건)-> '열'에 있는 데이터가 '조건'을 만족하도록 대체.
#조건은 딕셔너리 활용: {'Male':'남', 'Female' :'여'}
tips['sex'] = tips['sex'].replace({'Male':'남', 'Female' :'여'})
tips


#3. size가 4 이상인 경우 '단체 손님, 이하인 경우 '-'을 적용해 tips['type']열에 넣어주기

def customer_type(x):
    if x >= 4:
        return '단체손님'
    else:
        return '-'


tips['type'] = tips['size'].apply(customer_type)
tips
```

```python
# 1. 팁/전체 계산 금액 으로 전체 금액 대비 팁의 비율 계산해 tips['ratio']열 생성
#함수냐 for 문이냐 고민이 되는구먼.

#문제점: 행/행 결과가 뭉텅이로 리스트에 넣어서 그럼
#해결점: 위치로 잡아서 리스트로 넣어줘야 원하는 결과가 나올것

ratio =[]

for i in range(len(tips)):
    calcul = tips.loc[i ,'tip'] / tips.loc[i, 'total_bill']
    print(calcul)
ratio.append(calcul)


#len(tips)== len(calcul)

#df['ratio'] = ratio     #여기 오류난다 (칼럼 길이 불일치)

ratio = pd.DataFrame(ratio)     #이렇게 하면 열에 들어가느냐? 몰루 이게 무슨 작업이지 데프화 아닌가 알아냈다!!!(데프화 맞어!!)

result = pd.concat([tips,ratio],axis = 1) #ratio 열 생성

result
```
