### bar chart (막대 그래프)

- 막대
- 히스토그램(도수분포표)

```python
#plt.subplots(행, 열, figsize=(가로, 세로))
# fig = 도화지 한 장, ax = 그림. 그림 몇 장, 도화지 사이즈
#지금은 행, 열 옵션을 주지 않았으니 1행, 1열 => 1*1 => 1개짜리 그림
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'green', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']

#그림. 막대 (행은 과일, 열은 카운트, 바 범례는 바 레이블, 컬러는 바 컬러)
# => 과일별 갯수(~counts까지)
# bar_labels 옵션은 막대 차트만으로는 안나옴 ax.legend 가 필요 (범례니까)
# color 옵션은 막대 색 지정 (hex code 사용시 #~). 정해진 x축 길이만큼 []로 묶어 전달. #색 통일시 꼭 길이만큼 중복 작성하지 않아도 됨
                         # 단 한 가지만 강조하고 2지선다 색 적용하고 싶으면 강조되지 않는 길이만큼 반복되는 색을 길이만큼 작성해야함.
ax.bar(fruits, counts, label = bar_labels, color = bar_colors)
ax.legend()             #bar_labels = ['red', 'blue', '_red', 'orange']에서 _red는 무시되어 범례에는 3색만 나옴


#제목
ax.set_title ('Fruits Amount')

#축 제목
ax.set_ylabel('supply', rotation = 'horizontal')             #축 제목 로테이션 옵션: rotation = 'horizontal' 덧붙이기
ax.set_xlabel('demand')


#그림 인쇄 (print 같은)
#plt.show()  #(한 셀 내 두 개의 그림이 나오는 경우 결과 살려서 보여주기용)

#그림 저장
#plt.savefig('path')
plt.savefig('/content/fruit.png')
```

### 상관관계 히트맵 그리기
 - 상관관계: x1, x2, x3 ... x끼리의 관계를 표현함

```python
#이중for문

for i in range(7):
    for v in vegetables:        # 첫 for 문의 원소(숫자), 내부의 원소(vegetables) => 0, 채소1, 0, 채소2... 밖과 안의 for 순서 중요합니다
        print(i, v)
```

```python
#상관관계 그래프 만들기

#y축
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]

#x축
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

#실제 히트맵에 들어갈 데이터 (더미)
harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

#1행 1열
fig,ax = plt.subplots()

image = ax.imshow(harvest)  # harvest는 이미 데이터가 있으니 이미지로 보여주세요~
#여기선 밝을수록 값이 큼

#set_xticks: x축 데이터명 고치기 (어느 자리에, 어떤 이름이) => 우리가 하고 싶은건 0에 joe~
ax.set_xticks(range(len(farmers)), labels = farmers, rotation = 45)            #len(farmers) = 7 / range(len(farmers)) = 0~6까지의 자리에

#옵션: rotation = 각도

#set_yticks: y축 데이터명 고치기
ax.set_yticks(range(len(vegetables)), labels = vegetables, rotation = 45)


#칸 안에 데이터 수치 나타내기 : 이중 for 문 사용     # 첫 for 문의 원소(vegetables), 내부의 원소(farmers) => 채소1, 0, 채소1, 1 ...
for v in range(len(vegetables)):                     # 첫 for 문: vegetables의 길이만큼 돕니다 (행 방향)
    for p in range(len(farmers)):                      #안의 for문:  행 방향 값 하나 당 farmers 길이만큼 돕니다  (열 방향)          즉, 채소 고정 until farmers 다 돌 때까지
        text = ax.text(p,v,harvest[v,p], ha = 'center', va = 'center', color = 'white')             # text는 (x,y) 사람(행), 채소(열), /여기까지 위치 잡고 이곳에 / 해당되는 수확 [채소v,사람p]값 넣어주세요


        #horizontal a~, vertical a(얼라인///align?) = 'center'옵션: 수평기준, 수직 기준 글씨 가운데 정렬
```

## 박스 플롯
- 이상치(IQR 방식)
 - 박스(전체 데이터 25%, 50%, 75% 의미)
 - 수염 (정상 범위)

if 정상범위를 벗어난 데이터가 있을 경우 > 동그라미로 표시

```python
np.random.seed(2026)                                        #랜덤 고정 뽑기 방법: 랜덤 데이터 뽑을 때 같은 조합(괄호 내 수)으로 적용. (for 연구 논문) np.random.seed(같은 수(대체로 42~))

#더미 데이터 생성
#np->넘파이, random.normal(평균, 표준편차, size = 개수) -> 평균으로부터 표준편차만큼 (평균, 표편의 위치는 고정) +- 된 사이에서 사이즈 만큼 정규분포 데이터 생성
fruits_weight = [np.random.normal(130, 10, size = 100),
                 np.random.normal(120, 10, size = 100),
                 np.random.normal(100, 30, size = 100)]

labels = ['apple', 'orange', 'tomato']



#그래프 그리기

fig, ax = plt.subplots()
#ax.boxplot(박스플롯 데이터, 각 축의 이름)
ax.boxplot(fruits_weight, tick_labels = labels)     #tick_labels :축 이름은 ~로 해주세용 옵션

plt.show()          #o표시는 이상치
```

## 히스토그램

```python
#가상 데이터
points = 100000  #데이터 개수
n_bins = 20      #구간

#standard_normal(데이터 개수): 데이터 개수만큼의 정규분포표 생성
dist1 = np.random.default_rng().standard_normal(points)     #랜덤으로,우선 레인지로 뽑고 그 안에서 정규 분포표


#그래프 그리기

fig, ax = plt.subplots()

#히스토그램 함수    hist(데이터, bins = 얼마나 쪼갤거니)
ax.hist(dist1, bins = n_bins)

plt.show()
```

## N행 M열 그래프 그리기

- 한 장의 fig에 여러 개의 ax를 적용하는 방법

```python
#더미 데이터
#arange = range와 동일. (시작, 끝(-1), 점프)
t = np.arange (0, 10, 0.01)

#그래프 그리기 ~(2행, 1열)
fig, ax = plt.subplots(2, 1)

#그래프가 2개 이상, = 리스트가 두개 이상 = 인덱싱 활용해

ax[0].plot(t,t)
ax[1].plot(t,np.exp(t))     #np.exp(지수)
plt.show()
```

```python
#더미 데이터
#arange = range와 동일. (시작, 끝(-1), 점프)
t = np.arange (0, 10, 0.01)

#그래프 그리기 ~(2행 2열)
fig, ax = plt.subplots(2, 2)

#그래프가 2개 이상, = 리스트가 두개 이상 = 인덱싱 활용해
#[행,열]로 그래프 위치값으로 부름
ax[0,0].plot(t,t, color = 'green')
ax[0,1].plot(t,np.exp(t), linestyle = '--')     #np.exp(지수)                                   #여기에 이중 for문 사용 가능
ax[1,0].scatter(t, t+3, s = 0.1)                                                               #plot을 다른거로 변경 가능
ax[1,1].plot(t,t*0.01)

plt.show()
```
