```python
import os
```

```python
os.listdir('/content/drive/MyDrive/새싹과정/my/seaborn 샘플 데이터')
#listdir(''): 폴더 경로를 복사해 붙이면 폴더 내 자료가 무엇이 있는지 되돌려줌 ('컴퓨터야,, 리스트를 읽어라 내 폴더의... 폴더 경로는 여기다...)
```

```python
origin = '/content/drive/MyDrive/새싹과정/my/seaborn 샘플 데이터'
```

```python
#리스트 활용
datas = []

for file in os.listdir(origin):        #for 이용해서 읽은 폴더 내에 csv 파일만 쏙쏙.
    if 'csv' in file:
                                                #os.path.join(앞에 올 경로, 뒤의 파일 이름)
        #print (os.path.join(origin, file))      #개별 파일을 절대경로로 가져오기

        datas.append(os.path.join(origin, file))        #리스트에 넣어주세용 리스트는 ()여기의 내용을 넣어주심 되용

datas[0]
```

```python
#딕셔너리 활용
datas = {}

for file in os.listdir(origin):        #for 이용해서 읽은 폴더 내에 csv 파일만 쏙쏙.
    if 'csv' in file:
                                                #os.path.join(앞에 올 경로, 뒤의 파일 이름)
        #print (os.path.join(origin, file))      #개별 파일을 절대경로로 가져오기

        datas[file.split('.')[0]] = os.path.join(origin, file)              #datas[file.split(',')[0]]: file (파일명). 분리해주세요 ',' 기준으로, [0]은 인덱싱? 첫번째 값만 가져와주세용 (결과는 '파일명'만 분리되서 가져와짐)
                                                                                # => 이는 datas[키], 이후로 = os.path.join(origin, file)  가 값에 해당하는 것. 결과 조회해보면 딕셔너리로 옴

datas
```
